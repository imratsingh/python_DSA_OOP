# instance variable 

```python
class student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
s1 = student("imrat",32)
s2 = student("lodhi",44)
print(s1.name,s1.age)
print(s2.name,s2.age)
```

# class variable

```python
class student:
    school = "ABC School"
    def __init__(self,name):
        self.name = name
s1 = student("ram")
s2 = student("shyam")
print(s1.school) # through object
print(student.school) # we can also colled class variable through class name
```

# local variable

```python
class test:
    def show(self):
        x = 10
        print(x)
t = test()
t.show()
```

# __init__ method

```python
class student:
    def __init__(self,name,age):
        self.name = name
        self.age = age
s1 = student("imrat",23)
s2 = student("imrat singh",45)
print(s1.name,s1.age)
print(s2.name,s2.age)
```

# instance method

```python
class student:
    def __init__(self,name):
        self.name = name
    def show(self):
        print("name",self.name)
s1 = student("ram")
s1.show()
```

# class method

```python
class student:
    school = "ABC SCHOOL"
    @classmethod
    def get_school(cls):
        print("scholl",cls.school)
student.get_school() # directly colled class method 
s1 = student() # colled class method with the help of object
print(s1.school)
```

# access or modify class variable

```python
class student:
    school = "ABC school"
    @classmethod
    def change_name_school(cls,new_school):
        cls.school = new_school

student.change_name_school("XYZ school")
print(student.school)
```

# static method

```python
class math:
    @staticmethod
    def add(a,b):
        return a+b
print(math.add(2,3))
```

```python
class student:
    school_name = "shsh"
    grade = "10th"
    def __init__(self,name,roll):
        self.n = name
        self.r = roll
    @classmethod
    def update_grad(cls,new_grad,city):
        cls.grad = new_grad
        cls.c = city
obj = student("nery",101)
print(student.school_name,obj.n,obj.r)
obj.update_grad("11","bhopal")
print(student.c)
```
