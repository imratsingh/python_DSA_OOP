# inheritance

-- single lavel inheritance

```python
class parent:
    x = 10
    def home(self):
        print("parent class")
class child(parent):
    pass
obj = child()
print(obj.x)
obj.home()
```

# multilavel inheritance

-- When a class inherits from another class, and then another class inherits from that child class, it is called Multilevel Inheritance.

```python
class A:
    def home(self):
        print("from grand parent")
class B(A):
    def home(self):
        super().home()
        print("from home P")
    def car(self):
        print("from car P")
class C(B):
    pass
obj = C()
obj.home()
obj.car()
```

# multipel inheritance

-- When a class inherits from more than one parent class, it is called Multiple Inheritance.

```python
class A:
    def home(self):
        print("home from A")
class B:
    def home(self):
        print("home from B")
        A().home()
    def car(self):
        print("car from B")
class C(B,A):
    pass

obj = C()
obj.home()
obj.car()
```

```python
class A:
    def home(self):
        print("home from A")
        B().home()
class B:
    def home(self):
        print("home from B")
         
    def car(self):
        print("car from B")
class C(A,B):
    pass

obj = C()
obj.home()
obj.car()
```

# harichal inheritance

-- When multiple child classes inherit from a single parent class, it is called Hierarchical Inheritance.

```python
class A:
    def home(self):
        print("from class A")
class B(A):
    def home(self):
        print("from class B")
class C(A):
    def home(self):
        print("from class C")

obj1 = B()
obj2 = C()
obj1.home()
obj2.home()
```

# hybrid inheritance

-- Hybrid inheritance is a combination of two or more types of inheritance in a single program.

```python
class A:
    def home(self):
        print("home from A")
class B(A):
    def home(self):
        super().home()
        print("home from B")
    def car(self):
        print("car from B")
class C(B):
    def school(self):
        print("school")
class D(C):
    def car1(self):
        print("from D class ")
obj1 = B()
obj2 = C()
obj3 = D()
obj3.home()
```

```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        print("B")

class C(A):
    def show(self):
        print("C")

class D(B, C):
    pass

obj = D()
obj.show()
```

```python
class calculater(ABC):
    def add(self,*n):
        sum = 0
        for i in n:
            sum = sum+ i
        print(sum)
    @abstractmethod
    def multi(self):
        pass
    @abstractmethod
    def divid(self):
        pass
class A(calculater):
    def multi(self,*n):
        multi = 1
        for i in n:
            multi = multi * i
        print(multi)
    def divid(self,*n):
        divid = 1
        for i in n:
            divid = divid / i
        print(divid)
obj = A()
obj.add(1,2,3,4,5)
obj.multi(1,2)
obj.divid(4,5)
```
