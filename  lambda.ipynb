```python
from functools import reduce
l = [1,2,3,4,5]
def add(sum,x):
    return sum+x
res= reduce(add,l)
print(res)
```

```python
from functools import reduce
l = [1,2,33,4,5]
def max(x,y):
    if x > y:
        return x
    else:
        return y
res = reduce(max,l)
print(res)
```

```python
from functools import reduce
l = [1,2,3,4,5,6]
def avg(x,y):
    return x+y
total = reduce(avg,l)
avg1 = total/len(l)
print(avg1)
```

# lambda function also know as anonomas fucntion

-- lambda(keyword) variable(n numberr of variable) : expression (single line expression)

-- it is not used return keyword

```python
var = lambda x,y:x+y
print(var(10,20))
```

```python
var = lambda x,y:x*y
varc = var(10,2)
print(varc)
```

```python
n=int(input("Enter your number :"))
var = lambda x:x**2
print(var(n))
```

-- lambda with if else

```python
n = int(input("Enter your number :"))
var = lambda x:"even" if x%2==0 else "odd"
print(var(n))
```

-- eval = it is the used when we wont input any types so it is used

```python
x = eval(input("Enter your number :"))
print(x,type(x))
```

--map with lambda

```python
l = eval(input("Etner your number :"))
var = list(map(lambda n:n**2,l))
print(var)
```

```python
l = eval(input("enter your number :"))
var = list(map(lambda n: n+5,l))
print(var)
```

```python
l1 = eval(input("enter your number :"))
l2 = eval(input("enter your number :"))
l3 = eval(input("enter your number :"))
var = list(map(lambda n1,n2,n3: n1+n2+n3,l1,l2,l3))
print(var)
```

-- filter with lambda

```python
l =[1,2,3,4]
res=list(filter(lambda n:n if n%2==0 else None , l))
print(res)
```

```python
l =[1,2,3,4]
res=list(map(lambda n:n if n%2==0 else "odd" , l))
print(res)
```
