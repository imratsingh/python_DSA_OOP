```python
def natural():
    n = 10
    i = 1
    while i <=n:
        print(i)
        i = i +1
res = natural()
print(res)
```

```python
def natural():
    n = 10
    i = 1
    while i <=n:
        yield (i)
        i = i +1
res = natural()
print(res)
for i in res:
    print(i)
```

```python
def natural():
    n = 10
    i = 1
    while i <=n:
        yield i
        i = i +1
res = natural()
print(res)
print(next(res))
print("hello")
print(next(res))
print("ehllo 2")
print(next(res))
```

# method overloding

```python
class math:
    def add(self,a,b=0,c=0):
        return a+b+c
M = math
print(M.add(2,3))
print(M.add(2,3,4))
```

# method overriding

```python
class animal:
    def sound(self):
        print("animal sound")
class dog(animal):
    def sound(self):
        print("dog barks")
d = dog()
print(d.sound())
d.sound()
```
